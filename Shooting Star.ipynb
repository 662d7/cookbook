```python
import yfinance as yf
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams["figure.figsize"] = [16, 10]
```

```python
df = yf.Ticker("^GSPC")
period = '300mo' # m = min, h = hour, d = day, wk = week, mo=month
interval = '1d' # valid intervals 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo
price = df.history(period=period, interval=interval)
```

```python
price
```

```python
bull = []
for ii in range(price.shape[0]):
    if (price['High'][ii] - price['Open'][ii])/-(price['Close'][ii]-price['Open'][ii]) >= 3:
        if price['Low'][ii]+(price['High'][ii]-price['Low'][ii])*0.15 - price['Close'][ii] > 0:
            if (price['High'][ii] - price['High'][ii-1]) > 0:
                if (price['High'][ii-1] - price['High'][ii-2]) > 0.:
                    if (price['High'][ii-2] - price['High'][ii-3]) > 0:
#             if (price['Open'][ii-1] - price['Close'][ii-1]) < 0:
#                 if (price['Open'][ii-2] - price['Close'][ii-2]) < 0:
#                     if (price['Open'][ii-3] - price['Close'][ii-3]) < 0:
                        bull.append(ii)
bull = np.array(bull)
bull = np.concatenate((bull, bull-1, bull-2, bull-3,bull+1, bull+2, bull+3,bull+4, bull+5, bull+6,
                       bull+7, bull+8, bull+9, bull+10, bull+11, bull+12))
bull = np.sort(bull)
```

```python
bull
```

```python
bear = []
for ii in range(price.shape[0]):
    if (price['High'][ii] - price['Close'][ii])/np.abs(price['Close'][ii]-price['Open'][ii]) >= 2:
        if (price['Open'][ii-1] - price['Close'][ii-1]) < 0:
            if (price['Open'][ii-2] - price['Close'][ii-2]) < 0:
                if (price['Open'][ii-3] - price['Close'][ii-3]) < 0:
                    bear.append(ii)
```

```python
fig = go.Figure(data=[go.Candlestick(x=price.index[bull],
                open=price['Open'][bull], high=price['High'][bull],
                low=price['Low'][bull], close=price['Close'][bull])
                     ])

fig.update_layout(xaxis_rangeslider_visible=False)

fig.update_layout(
    autosize=False,
    width=1000,
    height=600,
)
fig.show()
```
